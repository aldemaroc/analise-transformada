```python
import numpy as np
import matplotlib.pyplot as plt

# Como se trata de uma função real não há números imaginários, então podemos fazer u(n) direto na marra

n = np.arange(-10, 10) # valores limites
def u(n): # fazendo a função u(n)
    if n>=0:
        return 1
    else:
        return 0
vetor_u = np.vectorize(u)(n) # criando o vetor

#Gráficos

plt.stem(n,vetor_u)
plt.title('Função u(n) calculada')
plt.xlabel('n')
plt.ylabel('u(n)')
plt.grid()
plt.show()

theta=np.linspace(0, 2*np.pi, 100)
y=np.sin(theta)
x=np.cos(theta)
plt.figure()
plt.plot(x, y, '-', linewidth=1)
plt.axis('equal')
plt.axis([-2, 2, -2, 2])
plt.grid(True)
plt.ylim(-2, 2)
plt.xlim(-2, 2)
plt.axvline(0, lw=2, color='darkgray')
plt.axhline(0, lw=2, color='darkgray')

num=[1]
den=[1, -1]
zeros=np.roots(num)
polos=np.roots(den)
plt.plot(zeros.real, zeros.imag, 'o', markersize=10, label='Zeros')
plt.plot(polos.real, polos.imag, 'x', markersize=12, label='Polos')
plt.legend()
plt.show()

U = np.fft.fft(vetor_u)
plt.stem(n, np.abs(U))
plt.grid()
plt.title("Fourier da funcao")
plt.ylabel("Modulo")
plt.xlabel("n")
plt.show()

```
